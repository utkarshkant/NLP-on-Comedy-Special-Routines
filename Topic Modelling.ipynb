## 2. Topic Modelling: Nouns
- it is a standard practice to create a model on a single part of speech (i.e. only nouns or only adjectives etc.)
- Here we will consider all the nouns for LDA model


```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# let us create a function to pull out nouns from a string of text
from nltk import word_tokenize, pos_tag

# UDF to tokenize text and extract all nouns
def nouns(text):
    is_noun = lambda pos: pos[:2] == 'NN'
    tokenized = word_tokenize(text)
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)]
    return " ".join(all_nouns)
```

```python
# read in the cleaned data
data_clean = pd.read_pickle("clean_corpus.pkl")

data_clean
```

```python
# apply nouns UDF to clean data
data_noun = pd.DataFrame(data_clean.transcript.apply(nouns))
data_noun
```

```python
# imports for creating a new document term matrix
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# additional stop-words
add_stop_words = ['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# create document term matrix
cvn = CountVectorizer(stop_words=stop_words)
data_cvn = cvn.fit_transform(data_noun.transcript)
data_dtm = pd.DataFrame(data_cvn.toarray(), columns=cvn.get_feature_names())
data_dtm.index = data_noun.index
data_dtm
```

```python
# create a gensim corpus
corpusn = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtm.transpose()))

# create a vocabulary dictionary
id2wordn = dict((v,k) for k, v in cvn.vocabulary_.items())
```

```python
# let's model with 2 stop words
ldan = models.LdaModel(corpus=corpusn, num_topics=2, id2word=id2wordn, passes=10)
ldan.print_topics()
```

```python
# let's model with 3 stop words
ldan = models.LdaModel(corpus=corpusn, num_topics=3, id2word=id2wordn, passes=10)
ldan.print_topics()
```

```python
# let's model with 4 stop words
ldan = models.LdaModel(corpus=corpusn, num_topics=4, id2word=id2wordn, passes=10)
ldan.print_topics()
```

## 3. Topic Modelling: Nouns & Adjectives
- Here we will consider all the nouns & adjectives for LDA model
- fine tune our LDA model


```python
# UDF to tokenize text and extract all nouns & adjectives
def nouns_adj(text):
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    all_nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)]
    return " ".join(all_nouns_adj)
```

```python
# apply noun-adj UDF 
data_noun_adj = pd.DataFrame(data_clean.transcript.apply(nouns_adj))
data_noun_adj
```

```python
'''
# imports for creating a new document term matrix
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

# additional stop-words
add_stop_words = ['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said']
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)
'''

# create document term matrix
cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_noun_adj.transcript)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names())
data_dtmna.index = data_noun_adj.index
data_dtmna
```

```python
# create a gensim corpus
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

# create a vocabulary dictionary
id2wordna = dict((v,k) for k, v in cvna.vocabulary_.items())
```

```python
# let's model with 2 stop words
ldana = models.LdaModel(corpus=corpusn, num_topics=2, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
# let's model with 3 stop words
ldana = models.LdaModel(corpus=corpusn, num_topics=3, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
# let's model with 4 stop words
ldana = models.LdaModel(corpus=corpusn, num_topics=4, id2word=id2wordna, passes=10)
ldana.print_topics()
```

## 4. Identify Topics in Each Document

```python
# let's model with 4 stop words
ldana = models.LdaModel(corpus=corpusn, num_topics=4, id2word=id2wordna, passes=80)
ldana.print_topics()
```
